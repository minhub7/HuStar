```python
#HW7 mySet 클래스 구현
class mySet():
    def __init__(self, L=[]): #초기화 함수
        self.Set = []
        for i in L:
            if i not in self.Set: #self.Set 리스트에 없을 경우 요소 추가
                self.Set.append(i)
     
    def add(self, elem): #add 함수
        if elem not in self.Set: #self.Set에 elem가 없을 시 추가
            self.Set.append(elem)
            
    def discard(self, elem): #discard
        if elem in self.Set: #self.Set에 elem가 있을 경우 삭제
            self.Set.remove(elem)
    
    def clear(self): #clear
        self.Set.clear() #Set 초기화, 모든 원소 삭제
    
    def __len__(self): #len()
        length = 0
        for _ in self.Set: #self.Set을 탐색하며 length 측정해서 반환
            length += 1
        return length
    
    def __str__(self): #print() 출력
        return f"{{{', '.join(map(str, self.Set))}}}"
        
    
    def __contains__(self, elem): # in 함수
        if elem in self.Set: return True #self.Set에 elem이 있으면 True
        else: return False
        
    def __le__(self, other): # less equal, self <= other
        cnt = 0
        for idx in self.Set:
            if idx in other:
                cnt += 1
        if cnt == len(self.Set):
            return True
        return False
        
    def __ge__(self, other):
        cnt = 0
        for idx in other.Set:
            if idx in self.Set:
                cnt += 1
        if cnt == len(other):
            return True
        return False
    
    def __and__(self, other):
        res = []
        for idx in self.Set:
            if idx in other.Set:
                res.append(idx)
        return mySet(res)
    
    def __iand__(self, other):
        res = []
        for idx in self.Set:
            if idx not in other:
                res.append(idx)
        for i in res:
            self.Set.remove(i)
        return self
    
    def __or__(self, other):
        res = self.Set
        for idx in other.Set:
            if idx not in self.Set:
                res.append(idx)
        return mySet(res)
    
    def __ior__(self, other):
        for idx in other.Set:
            if idx not in self.Set:
                self.Set.append(idx)
        return self
    
    def __sub__(self, other):
        res = self.Set
        for idx in other.Set:
            if idx in self.Set:
                res.remove(idx)
        return mySet(res)

    
    def __isub__(self, other):
        for idx in other.Set:
            if idx in self.Set:
                self.Set.remove(idx)
        return self
    
    
    
def mainSet():
    a = mySet([1,2,3,4])
    b = mySet([1,2,3,4])
    
    print(a, b)    
    a.discard(4)
    b.discard(1)
    print("After discard: ", a, b); print()
    
    print(len(a))
    print(1 in a)
    print(1 in b)
    
    print(f"\na|b = {a|b}, a&b = {a&b}, a-b: {a-b}\n")
    print(f"a <= b: {a <= b}, a <= a | b: {a <= a | b}")
    print(f"a >= b: {a >= b}, a >= a & b: {a >= a & b}")
    print()
    
    b.clear(); print(b); print()
    
    c = mySet([1,2,3])
    d = mySet([3,4])
    address_c = id(c)
    c |= d
    print(c)
    print(address_c == id(c))
    print()
    
    c = mySet([1,2,3])
    d = mySet([3,4])
    address_c = id(c)
    c &= d
    print(c)
    print(address_c == id(c))
    print()
    
    c = mySet([1,2,3])
    d = mySet([3,4])
    address_c = id(c)
    c -= d
    print(c)
    print(address_c == id(c))
    print()
    
    
mainSet()
```
